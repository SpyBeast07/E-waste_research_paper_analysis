# Prediction Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
# === Step 1: Load CSV ===
df = pd.read_csv("final_merged_file.csv")
```

    RESHAPE DATA

It reshapes the table from wide format to long format.

This format is better for plotting, modeling, and analysis.

```python
# === Step 2: Reshape Data ===
df_long = df.melt(id_vars='Country', var_name='Year', value_name='Ewaste')
df_long['Year'] = df_long['Year'].astype(int)
df_long['Year_num'] = df_long['Year'] - df_long['Year'].min()
```

    LINEAR REGRESSION

Fits a straight line through your data.

Equation: y = a * x + b

Itâ€™s simple, fast, and works great if your data follows a clear upward or downward trend.

RÂ² score (lin_r2) tells how well the line fits the actual data (closer to 1 = better fit).

    
    POLYNOMIAL REGRESSION

Fits a curved line (parabola) through the data.

Equation: y = a*xÂ² + b*x + c

Useful if data has a non-linear pattern, like a sudden spike, curve, or dip.

Again, RÂ² score (poly_r2) shows how well this curved line fits.

    SO WHY BOTH??

**You compare their RÂ² scores:**

If the polynomial model gives a clearly better fit (by 1% or more), you use that.

Otherwise, you stick to the simpler linear model.

    Automatically choose the best-fitting trend line for each country.

    Improve prediction accuracy.

```python
# === Step 3: Predict for Next 10 Years ===
future_years = list(range(2025, 2035))
future_nums = [y - 2017 for y in future_years]  # Normalize: 2017 â†’ 0

for country in df['Country']:
    data = df_long[df_long['Country'] == country]
    X = data[['Year_num']]
    y = data['Ewaste']

    # Linear Regression
    lin_model = LinearRegression().fit(X, y)
    lin_pred = lin_model.predict(X)
    lin_r2 = r2_score(y, lin_pred)

    # Polynomial Regression
    poly = PolynomialFeatures(degree=2)
    X_poly = poly.fit_transform(X)
    poly_model = LinearRegression().fit(X_poly, y)
    poly_pred = poly_model.predict(X_poly)
    poly_r2 = r2_score(y, poly_pred)

    # Choose best model
    if poly_r2 > lin_r2 + 0.01:
        model_type = "Polynomial Regression (deg=2)"
        X_future = poly.transform(np.array(future_nums).reshape(-1, 1))
        future_preds = poly_model.predict(X_future)
        fit_pred = poly_pred
    else:
        model_type = "Linear Regression"
        X_future = np.array(future_nums).reshape(-1, 1)
        future_preds = lin_model.predict(X_future)
        fit_pred = lin_pred

    # === Step 4: Statistical Summary ===
    avg_growth = np.mean(np.diff(y))
    std_dev = y.std()

    future_growth = np.mean(np.diff(future_preds))

    print(f"\nðŸ“Š --- {country} ---")
    print(f"Best Model Used: {model_type}")
    print(f"RÂ² Score: {max(poly_r2, lin_r2):.4f}")
    print(f"Average Annual Growth (2017â€“2024): {avg_growth:.0f} tons")
    print(f"Standard Deviation: {std_dev:.0f}")
    print(f"Projected Avg Growth (2025â€“2034): {future_growth:.0f} tons")
    
    print("\nðŸ“ˆ Predicted E-Waste (2025â€“2034):")
    for yr, val in zip(future_years, future_preds):
        print(f"{yr}: {val:,.0f} tons")

    # === Step 5: Plot Actual + Predictions ===
    plt.figure(figsize=(8, 4))
    plt.scatter(X, y, label='Actual (2017â€“2024)', color='blue')
    plt.plot(X, fit_pred, label='Model Fit', color='green')
    plt.plot(future_nums, future_preds, label='Future Prediction', color='red', linestyle='--')
    plt.xticks(ticks=range(0, 18), labels=range(2017, 2035), rotation=45)
    plt.xlabel("Year")
    plt.ylabel("E-waste (tons)")
    plt.title(f"{country} â€” E-Waste Prediction (2017â€“2034)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

**How much e-waste is expected to grow in each country from 2024 to 2034??**

```python
# === Step 6: Compare % Growth Across Countries ===
growth_stats = []

for country in df['Country']:
    data = df_long[df_long['Country'] == country]
    X = data[['Year_num']]
    y = data['Ewaste']

    # Train models again
    lin_model = LinearRegression().fit(X, y)
    lin_pred = lin_model.predict(X)
    lin_r2 = r2_score(y, lin_pred)

    poly = PolynomialFeatures(degree=2)
    X_poly = poly.fit_transform(X)
    poly_model = LinearRegression().fit(X_poly, y)
    poly_pred = poly_model.predict(X_poly)
    poly_r2 = r2_score(y, poly_pred)

    if poly_r2 > lin_r2 + 0.01:
        X_future = poly.transform(np.array([17]).reshape(-1, 1))  # 2034 = 2017 + 17
        pred_2034 = poly_model.predict(X_future)[0]
        model_type = 'Polynomial'
    else:
        X_future = np.array([[17]])
        pred_2034 = lin_model.predict(X_future)[0]
        model_type = 'Linear'

    ewaste_2024 = data[data['Year'] == 2024]['Ewaste'].values[0]
    percent_growth = ((pred_2034 - ewaste_2024) / ewaste_2024) * 100

    growth_stats.append({
        'Country': country,
        'Model': model_type,
        '2024 (Actual)': int(ewaste_2024),
        '2034 (Predicted)': int(pred_2034),
        '% Growth (2024â€“2034)': round(percent_growth, 2)
    })
```

```python
# Convert to DataFrame and sort
growth_df = pd.DataFrame(growth_stats)
growth_df = growth_df.sort_values(by='% Growth (2024â€“2034)', ascending=False)
```

```python
# === Step 7: Show Table & Plot ===
print("\nðŸ“Š Country-wise % Growth from 2024 to 2034:\n")
print(growth_df.to_string(index=False))
```

```python
# Bar Chart
plt.figure(figsize=(10, 5))
plt.bar(growth_df['Country'], growth_df['% Growth (2024â€“2034)'], color='orange')
plt.xticks(rotation=45)
plt.title("% E-Waste Growth by Country (2024â€“2034)")
plt.ylabel("% Growth")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
